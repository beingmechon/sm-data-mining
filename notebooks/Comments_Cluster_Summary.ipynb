```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import json


# Colab
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.headless = True
driver = webdriver.Chrome(options=chrome_options)

# Set up the Chrome WebDriver
# driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))

# Open the YouTube video
video_url = 'https://www.youtube.com/watch?v=avffSSsVklU'
driver.get(video_url)

# Wait for the page to load
WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.TAG_NAME, "body")))

# Scroll to load comments
def scroll_to_load_comments():
    last_height = driver.execute_script("return document.documentElement.scrollHeight")
    scroll_attempts = 0

    while True:
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")

        # Wait for comments to load
        time.sleep(5)

        new_height = driver.execute_script("return document.documentElement.scrollHeight")

        if new_height == last_height:
            scroll_attempts += 1
            if scroll_attempts > 5:
                break
        else:
            scroll_attempts = 0

        last_height = new_height

# Scroll to load all comments
scroll_to_load_comments()

# Additional wait time to ensure all comments are loaded
time.sleep(10)

# Scrape comments
try:
    comments = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, '//*[@id="content-text"]'))
    )
    comment_list = [comment.text for comment in comments]
except TimeoutException:
    comment_list = []

# Save the HTML content of the page
html_content = driver.page_source

# Close the browser
driver.quit()

# Save comments to a JSON file
video_id = video_url.split('=')[1]
with open(f'comments_{video_id}.json', 'w', encoding='utf-8') as f:
    json.dump(comment_list, f, ensure_ascii=False, indent=4)

# Save the HTML content to a file
with open(f'page_content_{video_id}.html', 'w', encoding='utf-8') as f:
    f.write(html_content)

print(f"Comments have been saved to comments_{video_id}.json")
print(f"HTML content has been saved to page_content_{video_id}.html")
```

```python
from sklearn.cluster import AgglomerativeClustering
from sentence_transformers import SentenceTransformer
import numpy as np
from bertopic import BERTopic
import json
from transformers import BartTokenizer, BartForConditionalGeneration

def get_similarity_matrix(sentences, similarity_function='COSINE'):
    """
    Compute the similarity matrix for a list of sentences using SentenceTransformer.

    Parameters:
        sentences (list): List of sentences.
        similarity_function (str): Name of the similarity function. Default is "COSINE".
                                   Other options are "DOT_PRODUCT", "EUCLIDEAN", and "MANHATTAN".

    Returns:
        numpy.ndarray: Similarity matrix.
    """
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    embeddings = model.encode(sentences)
    return np.matmul(embeddings, embeddings.T) if similarity_function == 'COSINE' else embeddings

def get_cluster_labels(similarity_matrix, threshold=0.7, n_clusters=None, metric='precomputed', linkage='average'):
    """
    Perform hierarchical clustering on the similarity matrix.

    Parameters:
        similarity_matrix (numpy.ndarray): Similarity matrix.
        threshold (float): Threshold distance for forming clusters. Default is 0.7
        n_clusters (int): Number of clusters. Default is None.
        metric (str): Affinity type. Default is 'precomputed'.
        linkage (str): Linkage criterion. Default is 'average'.

    Returns:
        numpy.ndarray: Cluster labels.
    """
    clustering = AgglomerativeClustering(n_clusters=n_clusters, metric=metric, linkage=linkage, distance_threshold=threshold)
    cluster_labels = clustering.fit_predict(1 - similarity_matrix)
    return cluster_labels

# def generate_cluster_summaries(clusters, sentences):
#     """
#     Generate concise summaries for each cluster using BART model.

#     Parameters:
#         clusters (dict): Dictionary with cluster_id as key and list of sentence indices as values.
#         sentences (list): List of original sentences.

#     Returns:
#         dict: Dictionary with cluster_id as key and concise summary as value.
#     """
#     tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
#     model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

#     cluster_summaries = {}
#     for cluster_id, indices in clusters.items():
#         selected_sentences = [sentences[idx] for idx in indices]
#         input_text = " ".join(selected_sentences)

#         inputs = tokenizer(input_text, return_tensors='pt', max_length=1024, truncation=True)
#         summary_ids = model.generate(inputs['input_ids'], num_beams=4, min_length=30, max_length=150, length_penalty=2.0, early_stopping=True)
#         summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

#         cluster_summaries[cluster_id] = summary.strip()

#     return cluster_summaries

# similarity_matrix = get_similarity_matrix(comment_list)

# cluster_labels = get_cluster_labels(similarity_matrix, threshold=0.7)

# clusters = {}
# for i, label in enumerate(cluster_labels):
#     str_label = str(label)
#     if str_label not in clusters:
#         clusters[str_label] = []
#     clusters[str_label].append(comment_list[i])


# print(clusters)
# cluster_summaries = generate_cluster_summaries(clusters, comment_list)

# output = {"cluster_summaries": cluster_summaries}

# with open('comment_cluster_summaries.json', 'w', encoding='utf-8') as f:
#     json.dump(output, f, ensure_ascii=False, indent=4)

# print("Cluster summaries have been saved to comment_cluster_summaries.json")
```

```python
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
```

```python
from transformers import pipeline


def generate_cluster_summaries(clusters, sentences, model_name):
    """
    Generate concise summaries for each cluster using BART model.

    Parameters:
        clusters (dict): Dictionary with cluster_id as key and list of sentence indices as values.
        sentences (list): List of original sentences.

    Returns:
        dict: Dictionary with cluster_id as key and concise summary as value.
    """
    # tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
    # model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

    summarizer = pipeline("summarization", model=model_name)


    cluster_summaries = {}
    for cluster_id, sentences in clusters.items():
        # print(cluster_id, sentences)
        selected_sentences = [sentence for sentence in sentences]
        input_text = "".join(selected_sentences)
        input_length = len(input_text)
        char_limit = 50

        if input_length > char_limit:
          max_length = input_length if input_length < 150 else 150
          min_length = char_limit if input_length < char_limit else input_length

          inputs = tokenizer(input_text, return_tensors='pt', max_length=1024, truncation=True)
          # summary_ids = model.generate(inputs['input_ids'], num_beams=4, min_length=min_length, max_length=max_length, length_penalty=1, early_stopping=True)
          summary_ids = model.generate(inputs['input_ids'],
                              num_beams=4,
                              min_length=min_length,
                              max_length=max_length,
                              length_penalty=1,
                              early_stopping=False,
                              temperature=150,
                              repetition_penalty=0.5)
          summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

        # summary = summarizer(input_text, max_length=150, min_length=30, do_sample=False)[0]['summary_text']
          cluster_summaries[cluster_id] = summary.strip()
        #   # print(summary.strip())
        else:
          print(f"Ignoring summarization for comments clusters with less than {char_limit} characters.")
          cluster_summaries[cluster_id] = input_text
        #   # print(input_text)

    return cluster_summaries
```

```python
import googleapiclient.discovery
from googleapiclient.errors import HttpError

def fetch_from_api(api_key, video_id="eUNDOgBiUds", page_token=None):
    api_service_name = "youtube"
    api_version = "v3"

    youtube = googleapiclient.discovery.build(
        api_service_name, api_version, developerKey=api_key)

    try:
        request = youtube.commentThreads().list(
            part="snippet",
            maxResults=100,
            order="relevance",
            videoId=video_id,
            pageToken=page_token
        )
        response = request.execute()
        return response

    except HttpError as e:
        print(f"An HTTP error {e.resp.status} occurred: {e.content}")
        return None

def fetch_comments_threads(api_key, video_id):
    raw_comments = []
    page_token = None

    while True:
        raw_data = fetch_from_api(api_key, video_id, page_token)
        
        if raw_data is None:
            break
        
        raw_comments.extend(raw_data.get("items", []))
        page_token = raw_data.get("nextPageToken")

        if not page_token:
            break

    return raw_comments


dev_key = ""
videoId = "sceqq4-kQX8"
raw_data = fetch_comments_threads(dev_key, videoId)

# for item in raw_data["items"]:
#     comments.append(item["snippet"]["topLevelComment"]["snippet"]["textOriginal"])
    
```
